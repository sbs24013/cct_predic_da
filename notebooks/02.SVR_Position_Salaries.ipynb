```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading the dataset

dataset = pd.read_csv("../datasets/50_Startups (2).csv")
dataset.head()
```

```python
# We split data into X and y

X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

# Reshape y
# print(y) [....]
y = y.reshape(len(y), 1) # [[],[],[],[],[]...,[]]
# print(y)
```

```python
#  We encode the categorical variable: State

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), [3])], remainder ='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
# Spitting the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# We can scale the features
from sklearn.preprocessing import StandardScaler
# https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html
# z = (x - u) / s

sc_X = StandardScaler()
sc_y = StandardScaler()
X_train = sc_X.fit_transform(X_train)
y_train = sc_y.fit_transform(y_train)
```

```python
# We start the model
from sklearn.svm import SVR # Regression

regressor = SVR(kernel='rbf')
regressor.fit(X_train, y_train.ravel()) 
# np_array.ravel() flattened array
```

```python
# We need to transform our data again

y_pred = regressor.predict(sc_X.transform(X_test)) 
# 훈련 데이터세트에 사용한 스케일을 사용하여 테스트 세트로 스케일한 뒤 모델에 전달
y_pred = y_pred.reshape(-1, 1)
# We reshape the data to be a 2D array so we can do later an interse transformation of the data
# transform your array into a 2D column vector
# [1, 2, 3, 4] -> [[1] [2] [3][4]] 
y_pred = sc_y.inverse_transform(y_pred)
```

```python
np.set_printoptions(precision=2) # 소수점 제한
print(
  np.concatenate(
    (
      y_pred,
      y_test.reshape(len(y_test), 1)
    ), 
    1
  )
)
# [[],[],[],...] -> [...]
```

```python
# Evaluate the model
from sklearn.metrics import r2_score

r2_score(y_test, y_pred)
```
