```python
#Importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = pd.read_csv('../datasets/groceries - groceries.csv')
data.head()
```

```python
# Convert the dataset into a list of lists (each list is a transaction)
transactions = data.astype(str).values.tolist()
```

```python
# Create a list of unique items
items = sorted(list(set(item for sublist in transactions for item in sublist if item != 'nan')))
```

```python
# Create a DataFrame where each column is an item and each row is a transaction (one-hot encoded)
encoded_data = pd.DataFrame([[item in transaction for item in items] for transaction in transactions], columns=items)
```

```python
# Apply the Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(encoded_data, min_support=0.01, use_colnames=True)
```

```python
# Generate association rules from the frequent itemsets
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
```

```python
# Display the results
from IPython.display import display

print("Frequent Itemsets:")
display(frequent_itemsets.head())
```

```python
print("\nAssociation Rules:")
display(rules.head())
```
